```python
# To specify PoS
doc = nlp("Elon flew to the mars yesterday. He carried biriyani masala with him.")

for token in doc:
  print(token, " | ", token.pos)
```

```python
doc = nlp("Elon flew to the mars yesterday. He carried biriyani masala with him.")

for token in doc:
  print(token, " | ", token.pos_)
```

```python
doc = nlp("Wow! Dr. Strange made 265 million $ on the very first day.")

for token in doc:
  print(token, " | ", token.pos_, " | ", spacy.explain(token.pos_), 
        " | ", token.tag_, " | ", spacy.explain(token.tag_))
# Here, we are using tag fucn to know more details about each parts of speech
```

```python
doc = nlp("He quits the job.")

doc[1]
print(doc[1].text, " | ", doc[1].tag_, " | ", spacy.explain(doc[1].tag_))
```

```python
doc = nlp("He quit the job.")

doc[1]
print(doc[1].text, " | ", doc[1].tag_, " | ", spacy.explain(doc[1].tag_))
```

## We can see from the above 2 code snippets is that __spacy__ of NLP knows how to distinguish between 'quits' & 'quit' when we are using tag_ fucn.

## So, __spaCy__ is smart enough to figure out what kind of tense it is.

```python
# https://www.microsoft.com/en-us/investor/earnings/fy-2023-q2/press-release-webcast
# let's do some NLP Analysis
earnings_text = """ Microsoft Corp. today announced the following results for the quarter ended December 31, 2022, as compared to the corresponding period of last fiscal year:

·        Revenue was $52.7 billion and increased 2%  

·        Operating income was $20.4 billion GAAP and $21.6 billion non-GAAP, and decreased 8% and 3%, respectively

·        Net income was $16.4 billion GAAP and $17.4 billion non-GAAP, and decreased 12% and 7%, respectively

·        Diluted earnings per share was $2.20 GAAP and $2.32 non-GAAP, and decreased 11% and 6%, respectively etc."""

```

```python
doc = nlp(earnings_text) # Converting it into doc
```

```python
for token in doc:
  print(token, " | ", token.pos_, " | ", spacy.explain(token.pos_))
```

## It is also detecting space = space
## And 'etc.' = other
## These are not useful so we need to remove them

```python
for token in doc:
  if token.pos_ in ["SPACE","X","PUNCT"]:
    print(token, " | ", token.pos_, " | ", spacy.explain(token.pos_))      
```

## We can see these are all unnecessary stuff. We need to get rid of all these

```python
for token in doc:
  if token.pos_ not in ["SPACE","X","PUNCT"]:
    print(token, " | ", token.pos_, " | ", spacy.explain(token.pos_))      
```

```python
doc = nlp(earnings_text)

filtered_tokens = []

for token in doc:
  if token.pos_ not in ["SPACE","X","PUNCT"]:
    filtered_tokens.append(token)
```

```python
filtered_tokens[:20]
```

## Now, we can see all that garbage has been cleaned out

## And now suppose if we want to extract how many nouns, verbs or other pos are present in that text, we can use a convenient API called __count_by__.

```python
count = doc.count_by(spacy.attrs.POS)
count
# attrs = attributes
```

## Here, it's showing POS 103 has 9 counts. Let's see what is 103.

```python
doc.vocab[103].text
# It's saying there are 9 spaces in the text
```

```python
doc.vocab[96].text
# It's saying there are 3 proper nouns in the text
```

```python
for k, v in count.items():
  print(doc.vocab[k].text, " | ", v)
```

## So, we can see that there are 9 spaces, 3 proper nouns, 29 nouns, 9 verbs and so on in the text.
